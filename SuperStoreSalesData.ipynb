```python
plt.hist(df['Quantity'], bins=14, edgecolor='black', alpha=0.7, color='tab:blue')
plt.title('Purchase Frequency over Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
##Check for missing values:
df.isnull().sum()
```

```python
##Found no null values in this dataset.

##Clean the data, removing any duplicates:
df.drop_duplicates(inplace=True)
```

```python
##Convert any date columns to proper datetime format:
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
##Basic Data Analysis Get basic statistics and inspect the data:
df.describe()
df.info()
```

```python
##Exploratory Data Analysis (EDA)
##1. Sales by Region
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
region_sales.plot(kind='bar', title='Sales by Region')
plt.show()
```

```python
##2. Sales Over Time 
##Plotting sales over time:
df['Year'] = df['Order Date'].dt.year
sales_over_time = df.groupby('Year')['Sales'].sum()
sales_over_time.plot(kind='line', title='Sales Over Time')
plt.show()
```

```python
##3. Sales by Product Category
product_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
product_sales.plot(kind='bar', title='Sales by Product Category')
plt.show()
```
